```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:/Users/noahs\Documents/Compting/NBA_STATS/jupyter/seasons_stats.csv", encoding="latin1")
df
```

```python
df = df.drop_duplicates()
df = df.drop_duplicates(subset=['Player'])
df
```

```python
df.info()
```

```python
df['PPG'] = df['PTS'] / df['G']
df['RPG'] = df['TRB'] / df['G']
df['APG'] = df['AST'] / df['G']
df['SPG'] = df['STL'] / df['G']
df['BPG'] = df['BLK'] / df['G']
df['FTAPG'] = df['FTA'] / df['G']
averages_df = df[['Player', 'Year', 'PPG', 'APG', 'RPG', 'SPG', 'BPG', 'FG%', 'eFG%', '2P%', '3P%', 'FT%']].copy()
```

```python
averages_df
```

```python
totals_df = df[['Player', 'Year', 'PTS', 'AST', 'TRB', 'STL', 'BLK', 'FG', 'FGA', 'MP', 'G', '3P', '3PA', '2P', '2PA']].copy()
totals_df
```

```python
averages_df.to_csv("averages.csv", index = False)
totals_df.to_csv("totals.csv", index = False)
```

```python
league_averages_df = df.groupby('Year')[['PPG', 'APG', 'RPG', 'SPG', 'BPG', 'FG%', 'eFG%', '2P%', '3P%', 'FT%']].mean().reset_index()
```

```python
league_averages_df
```

```python
league_averages_df.to_csv("league_averages.csv", index = False)
```
